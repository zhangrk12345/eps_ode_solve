```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def ode(x, y, eps):
    u, du = y
    dudx = du
    d2udx2 = (abs(du) - 1) / eps
    return [dudx, d2udx2]

def target(s, eps):
    res = solve_ivp(lambda x,y: ode(x,y,eps), [-1, 1], [0, s], t_eval=[1])
    return res.y[0, -1]

def solve_bvp(eps):
    low, high = -5, 5
    for _ in range(80):
        mid = (low + high) / 2
        val = target(mid, eps)
        if val > 0:
            high = mid
        else:
            low = mid
    s_opt = (low + high) / 2
    x_grid = np.linspace(-1, 1, 400)
    sol_full = solve_ivp(lambda x,y: ode(x,y,eps), [-1,1], [0, s_opt], t_eval=x_grid)
    return sol_full, s_opt

eps_list = [0.5, 0.1, 0.01, 0.001]
x_ref = np.linspace(-1,1,400)
u_ref = 1 - np.abs(x_ref)

plt.figure(figsize=(10,5))
plt.plot(x_ref, u_ref, 'k--', label=r'极限解 $u=1-|x|$', lw=3)
for eps in eps_list:
    sol, s_opt = solve_bvp(eps)
    plt.plot(sol.t, sol.y[0], label=f'$\epsilon={eps}$')
plt.xlabel('$x$')
plt.ylabel('$u_\epsilon(x)$')
plt.title('消失粘性法：不同ε的解对比（收敛验证）')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

for eps in eps_list:
    _, s_opt = solve_bvp(eps)
    print(f'ε={eps}, u\'(-1) = {s_opt:.5f}')
```
