```python
import math
import numpy as np
import matplotlib.pyplot as plt

# 1. 计算阈值k对应的选中最优概率
def success_prob(N, k):
    total = 0.0
    for i in range(k, N):
        total += 1 / i
    return (k / N) * total

# 2. 遍历找最优阈值k*
def find_optimal_k(N):
    max_p = -1
    best_k = 0
    for k in range(0, N):
        p = success_prob(N, k)
        if p > max_p:
            max_p = p
            best_k = k
    return best_k, max_p

# 3. 测试不同总人数N
test_N = [10, 50, 100, 500, 1000]
print("===== 秘书问题最优策略计算结果 =====")
for N in test_N:
    k_opt, p_opt = find_optimal_k(N)
    ratio = k_opt / N
    print(f"总人数N={N}, 最优观望人数k={k_opt}, 占比={ratio:.4f}, 选中最优概率={p_opt:.4f}")

# 4. 绘制概率曲线，验证1/e最优
N_plot = 200
k_range = np.arange(0, N_plot)
prob_list = [success_prob(N_plot, k) for k in k_range]
ratio_list = k_range / N_plot

plt.figure(figsize=(9, 4))
plt.plot(ratio_list, prob_list, linewidth=2)
opt_t = 1 / math.e
plt.axvline(opt_t, color='red', linestyle='--', label=f'理论最优比例 1/e ≈ {opt_t:.4f}')
plt.xlabel('观望比例 k/N')
plt.ylabel('选中最优候选人的概率')
plt.title(f'N={N_plot} 不同观望比例的成功概率曲线')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
